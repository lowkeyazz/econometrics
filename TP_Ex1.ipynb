Q2 : Calculating the alpha risk

```python
#Function that returns alpha the probability of X < u_alpha


alpha = scipy.stats.norm.pdf((IC_nr[1]-mu)/(sigma/np.sqrt(n)))
print( "{:.3f}".format(alpha))
```

Q4 : Calculating the beta risk

```python
mu_1 = 394
```

```python
beta = scipy.stats.norm.cdf((IC_nr[1]-mu_1)/(sigma/np.sqrt(n))) - scipy.stats.norm.cdf((IC_nr[0]-mu_1)/(sigma/np.sqrt(n)))
print( "{:.2f}".format(beta))
```

```python
power = 1 - beta
print("{:.2f}".format(power))
```

Q6 : Multiple mu values and efficiency curve

```python
import matplotlib.pyplot as plt
```

```python
mu_values = np.linspace(380,420,41)
```

```python
def beta_function(mu,sigma=sigma,n=n,IC_nr=IC_nr):
    return scipy.stats.norm.cdf((IC_nr[1]-mu)/(sigma/np.sqrt(n))) - scipy.stats.norm.cdf((IC_nr[0]-mu)/(sigma/np.sqrt(n)))
```

```python
y_beta = [i for i in beta_function(mu_values)]
print(y_beta)
```

```python
plt.scatter(mu_values,y_beta)
plt.plot(mu_values,y_beta)
plt.show
```

```python
def power(mu,sigma=sigma,n=n,IC_nr=IC_nr):
    return 1 - beta_function(mu,sigma,n,IC_nr)
```

```python
y_power = [i for i in power(mu_values)]
print(y_power)
```

```python
plt.scatter(mu_values,y_power)
plt.plot(mu_values,y_power)
plt.show
```
