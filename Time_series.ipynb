```python
import numpy as np
from statsmodels.tsa.arima_process import arma_generate_sample
```

```python
import matplotlib.pyplot as plt
```

TP1 : Génération de données ARMA sous Python

```python
# Set the random seed for reproducibility
np.random.seed(123)
```

Pour la 1ere question, le modele est MA(1) avec coefficient -0.25

```python
# Generate the time series
ar_coefs = [1]
ma_coefs = [1,0.25]
n = 100
errors = np.random.normal(size=n)
time_series = arma_generate_sample(ar_coefs, ma_coefs, n, errors)
```

```python
plt.plot(time_series)
plt.title("Time Series Plot")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.show()
```

Pour la 1ere question, le modele est AR(2) avec coefficient 0,03 et 0,2

```python
# Generate the time series
ar_coefs = [1,-0.03,-0.2]
ma_coefs = [1]
n = 200
errors = np.random.normal(size=n)
time_series = arma_generate_sample(ar_coefs, ma_coefs, n, errors)
```

```python
plt.plot(time_series)
plt.title("Time Series Plot")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.show()
```

Pour la 1ere question, le modele est ARMA(1,2) avec coefficient AR de lag-1 0.2 et MA de lag-2 0.3 et 0.4

```python
# Generate the time series
ar_coefs = [1,-0.2]
ma_coefs = [1,-0.3,-0.4]
n = 1000
errors = np.random.normal(size=n)
time_series = arma_generate_sample(ar_coefs, ma_coefs, n, errors)
```

```python
plt.plot(time_series)
plt.title("Time Series Plot")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.show()
```

TP2: Décomposition d’une série temporelle sous Python

```python
import pandas
import quandl
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
QUANDL_KEY = 'heSfyoTJ7z6aoom3S1SR'
quandl.ApiConfig.api_key = QUANDL_KEY
df = quandl.get("WGC/GOLD_MONAVG_USD", start_date="2000-01-01", end_date="2011-12-31")
```

```python
df.rename(columns={'Value': 'price'}, inplace=True)
df = df.resample('M').last()
```

```python
decomposition_results = seasonal_decompose(df.price, model='multiplicative')
decomposition_results.plot().suptitle('Multiplicative Decomposition',fontsize=20)
```
